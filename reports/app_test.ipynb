```python
import os
import sys
sys.path.append(os.path.dirname(os.getcwd()))
from data.clean_data import neutor_weekday, neutor_last_week
import plotly.graph_objects as go
import plotly.express as px
import dash
from dash import Dash, dcc, html, Input, Output

df_neutor_weekday = neutor_weekday()
df_neutor_last_week = neutor_last_week()

# Initialize the app
app = dash.Dash(__name__)
app.config.suppress_callback_exceptions = True




app.layout = html.Div(
    children=[
        html.Div(className='row',
                 style={'display': 'flex'},
                 children=[
                    html.Div(className='four columns div-user-controls',
                             
                             children=[
                                html.H2('DASH -  Radverkehr Münster'),
                                html.P('Auswahl für die Fahrtrichtung.'),
                                html.Div([
                                    html.Div([
                                        dcc.Dropdown(
                                            options=[
                                                {'label': 'Neutor', 'value': 'Neutor'},
                                                {'label': 'Neutor EW', 'value': 'Neutor FR stadteinwärts'},
                                                {'label': 'Neutor AW', 'value': 'Neutor FR stadtauswärts'}
                                            ],
                                            value='Neutor',
                                            id='yaxis-column'
                                        )
                                    ])
                                ],
                                     style={'color': '#1E1E1E'}
                                     )
                                ],
                                style={'color': '#1E1E1E', 'width': '20%'},
                             ),
                    html.Div(className='eight columns div-for-charts bg-grey',
                             children=[
                                 dcc.Graph(id='boxplot_wd',
                                     config={'displayModeBar': False},
                                     animate=True),
                                 html.Div([
                                     dcc.Graph(
                                         figure=go.Figure(
                                             data=[
                                                 go.Bar(
                                                     name='SE',
                                                     x=df_neutor_last_week['Datum'],
                                                     y=df_neutor_last_week['Neutor FR stadteinwärts'],
                                                     offsetgroup=0
                                                 ),
                                                 go.Bar(
                                                     name='SA',
                                                     x=df_neutor_last_week['Datum'],
                                                     y=df_neutor_last_week['Neutor FR stadtauswärts'],
                                                     offsetgroup=0,
                                                     base=df_neutor_last_week['Neutor FR stadteinwärts']
                                                 )
                                             ],
                                             layout=go.Layout(
                                                showlegend=True,
                                                xaxis=dict(title='Datum'),
                                                yaxis=dict(title='Werte'),
                                                plot_bgcolor="black",
                                                colorway=["#5E0DAC", '#FF4F00', '#375CB1', '#FF7400', '#FFF400', '#FF0056'],
                                                template='plotly_dark',
                                                paper_bgcolor='rgba(0, 0, 0, 0)',
                                                margin={'t': 50}
                                             )
                                         )
                                     )
                                 ])
                            ],style={'width': '80%'})
                             
                            ])
        ]

)

    
@app.callback(
    Output('boxplot_wd', 'figure'),
    Input('yaxis-column', 'value'))
def update_figure(selected_column):
    fig = px.box(df_neutor_weekday, x='Wochentag', y=selected_column, points='all')
    fig.update_layout(
        plot_bgcolor="black",
        colorway=["#5E0DAC", '#FF4F00', '#375CB1', '#FF7400', '#FFF400', '#FF0056'],
        template='plotly_dark',
        paper_bgcolor='rgba(0, 0, 0, 0)',
        margin={'t': 50},
        height=500,
        autosize=True,
        #title={'text': 'Daily Change', 'font': {'color': 'white'}, 'x': 0.5},
        title="Boxplot für die Visualisierung des Verkehrs an den einzelnen Wochentagen",
    )
    return fig


if __name__ == '__main__':
    app.run_server(debug=False)
```
